### **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere a un estado en el que los animales no experimentan estrés por calor ni por frío, lo que les permite mantener su temperatura corporal central dentro de un rango óptimo con un mínimo esfuerzo metabólico. Cuando el ganado se desvía de esta zona de confort térmico, activa mecanismos fisiológicos y conductuales para intentar regular su temperatura, lo que puede afectar negativamente su salud, producción y bienestar.

#### **Factores que influyen en el confort térmico:**

1.  **Temperatura del aire (Temperatura de bulbo seco, $T_{bs}$):** Es el factor más obvio y directo. Temperaturas extremas, tanto altas como bajas, pueden causar estrés.
2.  **Humedad relativa (Temperatura de bulbo húmedo, $T_{bh}$):** La alta humedad dificulta la disipación de calor por evaporación (sudoración y jadeo), lo que es crítico para el ganado en ambientes cálidos.
3.  **Radiación solar:** La exposición directa al sol aumenta significativamente la carga de calor en los animales.
4.  **Velocidad del viento:** Un viento suave puede ayudar a disipar el calor corporal en ambientes cálidos, pero vientos fuertes en frío pueden aumentar la pérdida de calor.
5.  **Precipitación:** La lluvia puede enfriar a los animales, pero si se combina con bajas temperaturas y viento, puede aumentar el estrés por frío.

#### **Índices de Confort Térmico:**

Debido a la interacción compleja de estos factores, se han desarrollado índices de confort térmico para integrar varias variables meteorológicas en una sola métrica. El **Índice de Temperatura y Humedad (ITH)** es uno de los más utilizados, especialmente para ganado lechero. Evalúa la combinación de la temperatura de bulbo seco y la temperatura de bulbo húmedo para estimar la severidad del estrés por calor. Un ITH alto indica una mayor probabilidad de estrés térmico, lo que puede llevar a:

*   Disminución en el consumo de alimento.
*   Reducción en la producción de leche y carne.
*   Menor eficiencia reproductiva.
*   Supresión del sistema inmune.
*   Aumento de la susceptibilidad a enfermedades.
*   En casos severos, la muerte.

La identificación temprana de condiciones de estrés térmico mediante estos índices permite a los productores implementar estrategias de mitigación, como proporcionar sombra, aumentar la ventilación, modificar la dieta, o ajustar los horarios de pastoreo y ordeño, para salvaguardar la productividad y el bienestar animal.

#**Métodos y cálculos**


• Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72 * (𝑇_{𝑏𝑠}+𝑇_{𝑏ℎ}) + 40.6$$

Donde:

1. $T_{bs}: tempperatura bulbo seco (°C)$
2. $T_{bh}: Temperatura de bulbo húmedo (°C)$

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$  ambiente no estresante

2. Sí $ITH > 70 \space and \space ITH<= 78$ situación de alerta por estrés térmico

3. Sí $ITH > 78 \space and \space <= 83$ peligro por estrés térmico

4. Sí $ITH > 83$ situación crítica

````python
1. # ENTRADA: pedir tbs ("Temperatura de Bulbo Seco y tbh Temperatura de Bulbo Humedo")
tbs = float(input("Ingrese la temperatura bulbo seco (°C): "))
tbh = float(input("Ingrese la temperatura bulbo húmedo (°C): "))

# PROCESO: calculo de ith (Indice de Temperatura y Humedad)
ith = 0.72 * (tbs + tbh) + 40.6

#SALIDA: Imprimir Resultados, de la Evaluacion del ith y el Confort Termico
if ith <= 70:
  print("El ambiente no está estresante.")
elif 70 < ith <= 78:
  print("La situación de alerta por estrés térmico.")
elif 78 < ith <=83:
  print("El peligro por estrés térmico.")
else:
  print("La situación crítica.")

````

*   **`tbh = float(input("Ingrese la temperatura bulbo húmedo (°C): "))`**: Similar a la anterior, esta línea solicita y convierte la temperatura de bulbo húmedo.

```python
# PROCESO: calculo de ith (Indice de Temperatura y Humedad)
```

*   **`# PROCESO...`**: Este comentario marca la sección de procesamiento.
*   **`ith = 0.72 * (tbs + tbh) + 40.6`**: Aquí se calcula el Índice de Temperatura y Humedad (ITH) utilizando la fórmula proporcionada. Se suman la temperatura de bulbo seco (`tbs`) y la temperatura de bulbo húmedo (`tbh`), el resultado se multiplica por `0.72`, y finalmente se le suma `40.6`.
