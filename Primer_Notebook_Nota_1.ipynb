# **Aspectos basicos de la programación con python**

Un estudiante de ingeniería necesita **calcular el área de un terreno rectangular** para estimar la cantidad de material necesario para un proyecto agrícola.

La fórmula es:

[
Area = Largo \times Ancho
]

---

# 1. Definición del algoritmo

## Entradas

Son los datos que el usuario debe proporcionar:

* `largo` → longitud del terreno (metros)
* `ancho` → ancho del terreno (metros)

## Proceso

Operación matemática a realizar:

* Multiplicar el largo por el ancho

[
area = largo * ancho
]

Operadores usados:

* `*` → operador de **multiplicación**

## Salidas

Resultados que mostrará el programa:

* `area` → área del terreno en **metros cuadrados (m²)**

---

# 2. Pseudocódigo del algoritmo

```
INICIO

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

FIN
```

---

# 3. Código en Python

```python
# -----------------------------------------
# Cálculo del área de un terreno rectangular
# Autor: Estudiante de Ingeniería
# -----------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

### Un ejemplo aplicado a la Ingeniería

En lugar del clásico "Hola Mundo", veamos algo más útil. Imagina que necesitas calcular el **volumen de agua** necesario para regar un lote agrícola o para llenar un tanque de almacenamiento en una obra civil.

Para que vayamos practicando las **buenas prácticas**, analicemos este pequeño bloque de lógica:

````python
# Definición de una función para modularizar el cálculo
def calcular_volumen_cilindro(radio_m, altura_m):
    pi = 3.14159
    volumen = pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada con nombres descriptivos
radio_tanque = 2.5  # metros
altura_tanque = 5.0 # metros

# Llamada a la función
resultado_volumen = calcular_volumen_cilindro(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {resultado_volumen} metros cúbicos.")
````

#### ¿Qué notas en este código?

1. **Nombres descriptivos:** No usé `r` o `h`, sino `radio_m` y `altura_m` para saber qué representan y en qué unidades están.
2. **Modularización:** Usé una función (`def`) para que el cálculo sea reutilizable en cualquier parte del programa.
