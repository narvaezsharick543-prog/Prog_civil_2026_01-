#**Dosificación de Mezcla para Columnas**

**Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

Solicitar la altura de la columna en metros.

Solicitar el diámetro de la columna en metros.

Calcular el volumen de la columna: $$v = \pi * r^2 * h$$. (Sugerencia: usar 31416 o importar math.pi).

Calcular la cantidad de sacos: $$ sacos: v * 9.5$$

**.Prueba del programa** :


Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2):

Volumen ≈0.377 m3

Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).

````python
# ENTRADA:
#solicitar la altura de la columna en metros.

# Solicitar el diámetro de la columna en metros.

h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

#PROCESO:
#calcula el radio r , calcula el volumen de la columna, calcula sacos 4
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDAS: Imprime el resultado
print("Volumen ≈", round ( v , 3) , "m3")
print("Sacos necesarios ≈", round(sacos , 2 ))
````

```markdown
**Explicación de las líneas de código para el cálculo de los sacos de cemento:**

*   `r = d/2`:
    *   Esta línea calcula el radio (`r`) de la columna. El radio es la mitad del diámetro (`d`). Se divide el valor del diámetro ingresado por el usuario por 2.

*   `v = 3.1416 * r**2 * h`:
    *   Esta línea calcula el volumen (`v`) de la columna cilíndrica. Utiliza la fórmula matemática para el volumen de un cilindro: $$\pi * r^2 * h$$.
    *   `3.1416` es una aproximación del número pi (π).
    *   `r**2` calcula el radio elevado al cuadrado.
    *   `h` es la altura de la columna ingresada por el usuario.

*   `sacos = v * 9.5`:
    *   Esta línea calcula la cantidad total de sacos de cemento necesarios (`sacos`).
    *   Multiplica el volumen (`v`) calculado de la columna por `9.5`, que es la cantidad de sacos de cemento requeridos por cada metro cúbico de mezcla, según el planteamiento del problema.
```
